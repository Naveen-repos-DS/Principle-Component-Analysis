# Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import scale
```

```python
# Removing categorical column for effective components.
# Mean = 0 and standard deviation = 1
new_df = df.iloc[:,1:]

#Scaling dataset before PCA

std_df = scale(new_df)
```

```python
std_df
```

```python
# PCA

from sklearn.decomposition import PCA
```

```python
#Principle components(PCs)

pca = PCA()
pca_values = pca.fit_transform(std_df)
pca_values
```

```python
# The amount of variance that each PCA explains is

var = pca.explained_variance_ratio_
var
```

```python
var_1 = np.cumsum(np.round(var,4)*100)
var_1
```

```python
plt.plot(var_1,color='red')
plt.xlabel('Principle components')
plt.ylabel('Variance explained')
plt.show()
```

From the above observations, we can reduce the number of columns from 13 to 7. These 7 columns capture almost 90% of the information in the data

```python
pca = PCA(n_components=7)
x_pca = pca.fit_transform(std_df)
```

```python
# Eigen vectors

pca.components_
```

# Clustering with Original Data

```python
clust = df.copy()
clust_df = clust.iloc[:,1:]
clust_df
```

```python
from sklearn.preprocessing import scale

scale_data = scale(clust_df)
scale_data
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
  model = KMeans(n_clusters=i)
  cl = model.fit(scale_data)
  val = model.inertia_
  wcss.append(val)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

To interpret from Elbow curve, K = 3 will form optimal cluster because of the sudden drop from 2 to 3.

```python
opt_model = KMeans(n_clusters=3)
opt_model.fit(scale_data)
yhat = opt_model.predict(scale_data)
```

```python
yhat
```

```python
clusters = pd.DataFrame(yhat,columns=['Clusters'])
```

```python
clust_df = pd.concat([clust_df,clusters],axis=1);clust_df
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(clust_df,opt_model.labels_)
```

**The model is performing with 20% accuracy**

```python
pca_model # Data transformed into PCs with 7 (for optimal PCs with 90% accuracy)
```

## Implementation of K-Means Clustering on PCA Tranformed Dataset

```python
from sklearn.cluster import KMeans

wcss_pca = []

for i in range(1,11):
  temp_model = KMeans(n_clusters=i,random_state=7)
  temp_clust = temp_model.fit(pca_model)
  inertia = temp_model.inertia_
  wcss_pca.append(inertia)
wcss_pca

plt.plot(wcss_pca, color='green')
```

```python
wcss_pca
```

```python

# Selected 3rd cluster because of maximum drop than others.

pca_kmean = KMeans(n_clusters=3)
cluster_pca = pca_kmean.fit_predict(pca_model)
```

# Comparison and Analysis

1.	Compare the clustering results obtained from the original dataset and PCA-transformed data.

***Without PCA (original features)***

* Silhouette score 0.20 clusters are present but overlapping.

* Separation by a few variables (especially Proline and Magnesium) other features add noise.

***With PCA (on 7 PCs)***

Silhouette score 0.35 clusters are more compact and better separated.

Clusters are clearly distinguished along PC1 and PC2.

2.	Discuss any similarities or differences observed in the clustering results.

Similarities: In both cases (with and without PCA), 3 clusters are formed.

differences: Silhouette improves from 0.20 (original) to 0.35 (PCA) clusters become more compact and better separated with PCA.

3.	Reflect on the impact of dimensionality reduction on clustering performance.

* PCA reduced the original 13 variables to 7 principal components that capture most of the variance.

* After PCA, the silhouette score increased from 0.20 to 0.35, showing that clusters became more compact and better separated.

* By removing noise and redundant features, PCA helped K-means focus on the most informative directions (PC1, PC2), which improved the stability and interpretability of the clusters.

4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.

PCA improves clustering performance and reduces dimensionality, but at the cost of some interpretability(its harder to say, cluster has high Proline and low Magnesium).

Clustering on the original data is easier to interpret but may give poorer cluster quality.

# Conclusion and Insights

K-means was run with and without PCA (k=3) on the wine data.

Original data: silhouette 0.20 moderate cluster separation. clusters mainly differ in Proline and Magnesium.

With PCA (7 PCs): silhouette 0.35 better, clearer clusters, mainly separated along PC1 and PC2.

***Use PCA + clustering when:***

You have many correlated or noisy features.

You want better cluster quality and separation (like our case: silhouette improved from 0.20 to 0.35).

Slight loss of interpretability is acceptable.

***Use clustering on original data when:***

Interpretability is very important  you need to explain clusters in terms of original variables (e.g., high Proline, low Magnesium).

The number of features is small and not highly redundant.

You mainly care about understanding which raw features drive each cluster.
